```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# สร้าง DataFrame
data = {
    'Gold Price Increment': [-1, -4, 2, -3, 1, -2, 4, 5, 2],
    'Bit_coin Increment': [846, 424, 1560, 523, 1012, 663, 1624, 1882, 1263]
}

# data = {
#     'Gold Price Increment': [-1, -4, 2, -3, 1, -2, 4, 5, 2],
#     'Bit_coin Increment': [846, 424, 3000, 523, 1012, 663, 1624, 1882, 1263]
# }

df = pd.DataFrame(data)

# Standardize คอลัมน์ 'Bit_coin Increment'
scaler_y = StandardScaler()
df['Bit_coin Increment Standardized'] = scaler_y.fit_transform(df[['Bit_coin Increment']])

# แสดงผลลัพธ์
print("Original DataFrame:")
print(df[['Gold Price Increment', 'Bit_coin Increment']])

print("\nStandardized DataFrame:")
print(df[['Gold Price Increment', 'Bit_coin Increment Standardized']])
```

```python
import matplotlib.pyplot as plt

# แสดง Scatter plot
plt.scatter(df['Gold Price Increment'], df['Bit_coin Increment Standardized'])
plt.title('Scatter plot of Gold Price Increment vs. Bit_coin Increment Standardized')
plt.xlabel('Gold Price Increment')
plt.ylabel('Bit_coin Increment Standardized')
plt.grid(True)
plt.show()
```

```python
import numpy as np

# คำนวณค่าเฉลี่ยของ x และ y
mean_x = np.mean(df['Gold Price Increment'])
mean_y = np.mean(df['Bit_coin Increment Standardized'])

# คำนวณค่า slope (theta1)
numerator = np.sum((df['Gold Price Increment'] - mean_x) * (df['Bit_coin Increment Standardized'] - mean_y))
denominator = np.sum((df['Gold Price Increment'] - mean_x) ** 2)
theta1 = numerator / denominator

# คำนวณค่า y-intercept (theta0)
theta0 = mean_y - theta1 * mean_x

# แสดงผลลัพธ์
print(f"Slope (theta1): {theta1}")
print(f"Y-intercept (theta0): {theta0}")
```

```python
import matplotlib.pyplot as plt

# แสดง Scatter plot
plt.scatter(df['Gold Price Increment'], df['Bit_coin Increment Standardized'], label='Data Points')

# สร้างเส้นตรง Linear Regression
x_values = np.linspace(df['Gold Price Increment'].min(), df['Gold Price Increment'].max(), 100)
y_values = theta0 + theta1 * x_values

# แสดงเส้นตรง Linear Regression
plt.plot(x_values, y_values, color='red', label='Linear Regression')

# ตกแต่งกราฟ
plt.title('Linear Regression: Gold Price Increment vs. Bit_coin Increment Standardized')
plt.xlabel('Gold Price Increment')
plt.ylabel('Bit_coin Increment Standardized')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ทำนายค่า Bit_coin Increment Standardized สำหรับ Gold Price Increment = 3
x_input = 3
predicted_y = theta0 + theta1 * x_input

# แสดงผลลัพธ์
print(f"Predicted Bit_coin Increment Standardized for Gold Price Increment = 3: {predicted_y}")
```
